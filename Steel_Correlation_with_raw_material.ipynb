```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load data
file_path = "/content/Steel All Components JAn-Oct 2024 Prices.xlsx"  # Ensure the file is in your working directory
df = pd.read_excel(file_path)
```

```python
# Step 2: Clean column names
df.columns = df.columns.str.strip()

# Step 3: Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
```

```python
# Step 4: Plot pricing trends (INR only for consistency)
plt.figure(figsize=(16, 8))

cols_to_plot = [
    'Mild Steel Price (INR)', 'Stainless Steel Price (INR)',
    'Steel Scrap Price (INR)', 'Iron Price (INR)',
    'Mt Coke Price (INR)', 'Silicomanganese Price (INR)',
    'Chromium Price (INR)', 'Nickel Price (INR)', 'Molybdenum Price (INR)'
]

for col in cols_to_plot:
    plt.plot(df['Date'], df[col], label=col)

plt.title('Prices of Steel and Its Components Over Time (INR)')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 5: Compute Correlation Matrix
correlation_df = df[cols_to_plot].corr()

# Step 6: Plot Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_df, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix: Steel Prices vs Component Prices (INR)')
plt.tight_layout()
plt.show()
```

```python

# Step 7: Print Explanation

# Sort correlations with Mild Steel Price (INR)
mild_corr = correlation_df['Mild Steel Price (INR)'].sort_values(ascending=False)

# Sort correlations with Stainless Steel Price (INR)
stainless_corr = correlation_df['Stainless Steel Price (INR)'].sort_values(ascending=False)

# Display the results
print("🔍 Correlation of Other Variables with Mild Steel Price (INR):\n")
print(mild_corr, "\n")

print("🔍 Correlation of Other Variables with Stainless Steel Price (INR):\n")
print(stainless_corr)

```
